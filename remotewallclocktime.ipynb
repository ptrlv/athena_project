```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
%matplotlib inline
pd.set_option('display.max_rows',15)

from pandas import *
from pandas.io.parsers import read_csv
```

```python
df = read_csv('7f8b4850-fc30-448b-81c2-32f07abdb266.csv', index_col=False, nrows=9999)
```

```python
fig, ax = plt.subplots(figsize=(12, 9))
groups = df[df.duration<14400].groupby(['match_apf_queue'])
#ax.set(title='Distribution of write duration', xlabel='Duration (s)', ylabel='Count')

ax.margins(0.05)

for name, group in groups:
    ax.plot(group.duration, group.remotewallclocktime, marker='.', linestyle='', label=name, alpha=1)
ax.legend().set_visible(True)
```
